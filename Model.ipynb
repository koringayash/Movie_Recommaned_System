```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    if len(L)==0:
        L.append('Nothing')
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert2(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    if len(L)==0:
        L.append('Nothing')
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(convert2)
```

```python
movies['keywords']
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 10:            
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
def fetch_producer(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Producer':
            L.append(i['name'])
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['director'] = movies['crew'].apply(fetch_director)
```

```python
movies['producer'] = movies['crew'].apply(fetch_producer)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['genres'] = new_df['genres'].apply(stem)
new_df['keywords'] = new_df['keywords'].apply(stem)
new_df['cast'] = new_df['cast'].apply(stem)
new_df['director'] = new_df['director'].apply(stem)
new_df['producer'] = new_df['producer'].apply(stem)
```

```python
new_df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv1 = CountVectorizer(max_features = 5000,stop_words='english')
cv2 = CountVectorizer(max_features = 5000,stop_words='english')
cv3 = CountVectorizer(max_features = 5000,stop_words='english')
cv4 = CountVectorizer(max_features = 5000,stop_words='english')
cv5 = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
# vectors = cv.fit_transform(new_df['tags']).toarray()
vectors1 = cv1.fit_transform(new_df['genres']).toarray()
vectors2 = cv2.fit_transform(new_df['keywords']).toarray()
vectors3 = cv3.fit_transform(new_df['cast']).toarray()
vectors4 = cv4.fit_transform(new_df['director']).toarray()
vectors5 = cv5.fit_transform(new_df['producer']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity1 = cosine_similarity(vectors1)

def recommendOnGenres(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity1[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    movies = []
    for i in movies_list:
        movies.append(new_df.iloc[i[0]].title)
        
    return movies
```

```python
movie_name = 'Avatar'

genresW = 4
keywordsW = 3
castW = 3
directorW = 2
producerW = 1
```

```python
movieDictionary = {}
```

```python
def function(movieList,movieDictionary,weight):
    for movie in movieList:
        if movie not in movieDictionary:
            movieDictionary[movie] = 0
        movieDictionary[movie] = movieDictionary[movie] + weight
        
    return movieDictionary
```

```python
movieDictionary = function(recommendOnGenres(movie_name),movieDictionary,genresW)
movieDictionary = function(recommendOnKeywords(movie_name),movieDictionary,keywordsW)
movieDictionary = function(recommendOnCast(movie_name),movieDictionary,castW)
movieDictionary = function(recommendOnDirector(movie_name),movieDictionary,directorW)
movieDictionary = function(recommendOnProducer(movie_name),movieDictionary,producerW)
```

```python
movieDictionary = sorted(movieDictionary.items(), key=lambda x: x[1], reverse=True)
```
